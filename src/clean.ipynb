```python
import pandas as pd

raw_path = "../data/raw/stock_market_dataset.csv"
output_path = "../data/processed/cleaned_stock_market_data.csv"

df = pd.read_csv(raw_path, header=[0, 1, 2])

new_columns = []

for col in df.columns:
    ticker = col[0]
    price_type = col[1]
    third_level = col[2]

    if ticker == "Ticker" and price_type == "Price" and third_level == "Date":
        new_columns.append("Date")
    else:
        clean_ticker = (
            ticker.replace("^", "")
            .replace("=F", "")
            .replace("GC", "Gold")
            .replace("CL", "Oil")
            .replace("GSPC", "SP500")
            .replace("TNX", "Treasury10Y")
        )

        new_columns.append(f"{clean_ticker}_{price_type}")

df.columns = new_columns

df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

df = df.sort_values("Date")
df = df.drop_duplicates(subset=["Date"])
df = df.reset_index(drop=True)

df.to_csv(output_path, index=False)

print("Cleaned dataset saved to:", output_path)
print(df.head())
print(df.columns.tolist())
```

```python
import pandas as pd

df = pd.read_csv("../data/processed/cleaned_stock_market_data.csv")

df["Date"] = pd.to_datetime(df["Date"])

print("Shape:", df.shape)
print("\nDate range:")
print(df["Date"].min(), "to", df["Date"].max())

print("\nColumns:")
print(df.columns.tolist())

print("\nMissing values:")
print(df.isnull().sum())

print("\nDuplicate dates:")
print(df["Date"].duplicated().sum())

print("\nData types:")
print(df.dtypes)
```

```python
df = df.sort_values("Date")

# Fill missing values using previous valid value
df = df.ffill()

# If first rows still have missing values, fill using next valid value
df = df.bfill()

print(df.isnull().sum().sum())
```

```python
df.to_csv("../data/processed/validated_stock_market_data.csv", index=False)
```
