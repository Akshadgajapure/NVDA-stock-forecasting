```python
import pandas as pd

df = pd.read_csv("../data/processed/validated_stock_market_data.csv")

df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")

# Daily returns
df["NVDA_Return"] = df["NVDA_Close"].pct_change()


# Moving averages
df["NVDA_SMA_7"] = df["NVDA_Close"].rolling(window=7).mean()
df["NVDA_SMA_30"] = df["NVDA_Close"].rolling(window=30).mean()
df["NVDA_SMA_90"] = df["NVDA_Close"].rolling(window=90).mean()

# Exponential moving averages
df["NVDA_EMA_12"] = df["NVDA_Close"].ewm(span=12, adjust=False).mean()
df["NVDA_EMA_26"] = df["NVDA_Close"].ewm(span=26, adjust=False).mean()

# MACD
df["NVDA_MACD"] = df["NVDA_EMA_12"] - df["NVDA_EMA_26"]
df["NVDA_MACD_Signal"] = df["NVDA_MACD"].ewm(span=9, adjust=False).mean()

# Rolling volatility
df["NVDA_Volatility_7"] = df["NVDA_Return"].rolling(window=7).std()
df["NVDA_Volatility_30"] = df["NVDA_Return"].rolling(window=30).std()

# Lag features
df["NVDA_Close_Lag_1"] = df["NVDA_Close"].shift(1)
df["NVDA_Close_Lag_7"] = df["NVDA_Close"].shift(7)
df["NVDA_Close_Lag_30"] = df["NVDA_Close"].shift(30)

# Target: next day close price
df["Target_Next_Close"] = df["NVDA_Close"].shift(-1)

# Drop rows created as NaN because of rolling/lag/target
df = df.dropna()

df.to_csv("../data/processed/model_ready_dataset.csv", index=False)

print("Model-ready dataset saved.")
print(df.shape)
print(df.head())
```
