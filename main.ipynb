```python
import yfinance as yf

tickers = {
    "NVDA": "NVIDIA",
    "^GSPC": "SP500",
    "^VIX": "VIX",
    "GC=F": "Gold",
    "CL=F": "Crude_Oil",
    "^TNX": "Treasury_10Y"
}

data = yf.download(
    list(tickers.keys()),
    start="2010-01-01",
    end="2026-01-01",
    interval="1d",
    group_by="ticker"
)

data.to_csv("stock_market_dataset.csv")
print(data.head())
```
